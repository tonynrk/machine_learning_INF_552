#### b.ii) Calculate the exact match and hamming loss with all columns

```python
def exact_match_n_hamming_loss(y,pred,flag):
    tmp = pd.DataFrame()
    if flag == "train":
        tmp[0] = pred[0][0]
        tmp[1] = pred[1][0]
        tmp[2] = pred[2][0]
    else:
        tmp[0] = pred[0][1]
        tmp[1] = pred[1][1]
        tmp[2] = pred[2][1]
    total_exact_match = 0
    total_hamming_loss = 0
    for i in range(y.shape[0]):
        total_hamming_loss += hamming_loss(y.iloc[i],tmp.iloc[i])
        total_exact_match += accuracy_score(y.iloc[i],tmp.iloc[i])
    total_exact_match = total_exact_match / y.shape[0]
    total_hamming_loss = total_hamming_loss / y.shape[0]
    return total_exact_match, total_hamming_loss
```

### Monte-Carlo Simulation: Perform the following procedures 50 times, and report the average and standard deviation of the 50 Hamming Distances that you calculate.

### (a) Use k-means clustering on the whole Anuran Calls (MFCCs) Data Set (do not split the data into train and test, as we are not performing supervised learning in this exercise). Choose k  {1, 2, . . . , 50} automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any other method you know.
### (b) In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species.
### (c) Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance, Hamming score, and Hamming loss5 between the true labels and the labels assigned by clusters.

#### Using Silhouettes

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from collections import defaultdict
```

```python
X = df_1.iloc[:,:-3]
y = df_1.iloc[:,[-3,-2,-1]]
```

```python
table_2 = defaultdict(list)
#iteration
for _ in range(1,51):
    #get best k by sillouette methods
    k,cluster_labels,sil_score = find_best_k_and_labels(range(2,51),X)
    table_2["Best k"].append(k)
    table_2["Silhouette Score"].append(sil_score)
    best_exact_match, best_hamming_loss = measures_k_means(k,cluster_labels,y)
    table_2["Exact Match"].append(best_exact_match)
    table_2["Hamming Score"].append(1-best_hamming_loss)
    table_2["Hamming Loss"].append(best_hamming_loss)
    table_2["Hamming Distance"].append(best_hamming_loss*3)
```

```python
def find_best_k_and_labels(range_n_clusters,X):
    best_avg = -2
    best_cluster = 0
    best_cl_labels = []
    for n_clusters in range_n_clusters:
        clusterer = KMeans(n_clusters=n_clusters)
        cluster_labels = clusterer.fit_predict(X)
        silhouette_avg = silhouette_score(X, cluster_labels)
        if best_avg < silhouette_avg:
            best_avg = silhouette_avg
            best_cluster = n_clusters
            best_cl_labels = cluster_labels
    return best_cluster, best_cl_labels, best_avg
```

```python
def clustering_exact_match_n_hamming_loss(y,pred):
    total_exact_match = 0
    total_hamming_loss = 0
    for i in range(y.shape[0]):
        total_hamming_loss += hamming_loss(y.iloc[i],pred.iloc[i])
        total_exact_match += accuracy_score(y.iloc[i],pred.iloc[i])
    total_exact_match = total_exact_match / y.shape[0]
    total_hamming_loss = total_hamming_loss / y.shape[0]
    return total_exact_match, total_hamming_loss
```

```python
def measures_k_means(k,cluster_labels,y):
    y_0 = y[["Family"]]
    y_0["pred_label_cluster"] = cluster_labels
    for i in range(k):
        new_label = y_0[y_0["pred_label_cluster"] == i].iloc[:,0].value_counts().index[0]
        y_0.loc[y_0["pred_label_cluster"] == i, "pred_label" ] = new_label
    y_1 = y[["Genus"]]
    y_1["pred_label_cluster"] = cluster_labels
    for i in range(k):
        new_label = y_1[y_1["pred_label_cluster"] == i].iloc[:,0].value_counts().index[0]
        y_1.loc[y_1["pred_label_cluster"] == i, "pred_label" ] = new_label
    y_2 = y[["Species"]]
    y_2["pred_label_cluster"] = cluster_labels
    for i in range(k):
        new_label = y_2[y_2["pred_label_cluster"] == i].iloc[:,0].value_counts().index[0]
        y_2.loc[y_2["pred_label_cluster"] == i, "pred_label" ] = new_label
    y_pred_label = pd.concat([y_0["pred_label"],y_1["pred_label"],y_2["pred_label"]], axis=1, sort=False)
    y_pred_label.columns = ["pred_family","pred_genus","pred_species"]
    best_exact_match,best_hamming_loss = clustering_exact_match_n_hamming_loss(y,y_pred_label)
    return best_exact_match,best_hamming_loss
```
